## Filtering Data

- Dibawah ini adalah filter data anak usia 12 tahun ke bawah dengan diagnosis terkena penyakit diabetes. Beberapa variabel pendukung yang memperkuat hasil diagnosis yaitu Kadar Gula Darah yang lebih dari normal yaitu > 126, Penanda Genetik, Kadar Insulin yang tidak normal (rendah) yaitu kurang dari 15, Autoantibodi yang poaitive sering dikaitkan dengan Diabetes Autoimun, dan BMI anak yang kurang dari 18,5 berarti underweight atau lebih dari 30 berarti obesitas. 

```python
a = data[
    (data['Age'] <= 12) &
    (data['Blood Glucose Levels'] > 126) &
    (data['Genetic Markers'] == 'Positive') &
    (data['Insulin Levels'] < 15) &
    (data['Autoantibodies'] == 'Positive') &
    ((data['BMI'] < 18.5) | (data['BMI'] > 30))
]
a
```

## Membuat Box Plot Distribusi Kadar Glukosa Darah pada Anak-Anak Sesuai Kriteria pada Filtering Data

```python
#Membuat Plot
plt.figure(figsize=(12, 6)) #Ukuran Figure
plt.boxplot([a['Blood Glucose Levels'][a['Genetic Markers'] == 'Positive']],
             labels=['Positive'])

# Judul dan Label Plot
plt.title('Boxplot of Blood Glucose Levels for Filtered Data')
plt.xlabel('Genetic Markers')
plt.ylabel('Blood Glucose Levels')

# Menampilkan plot
plt.show()
```

## Korelasi

```python
# Menghitung matriks korelasi
korelasi = a.corr()
print(korelasi)
```

```python
# Mengatur ukuran figure
plt.figure(figsize=(12, 10))

# Membuat heatmap
sns.heatmap(korelasi, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True)

# Menambahkan judul
plt.title('Korelasi antara Variabel dalam Dataset')

# Menampilkan plot
plt.show()
```
